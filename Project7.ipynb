1. Find the average study hours per day for all students. Please create a code cell below this to answer the question.(0.5 point)

```python
# Install dependencies as needed:
# pip install kagglehub[pandas-datasets]
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
# Update file_path to point to the specific file within the dataset
file_path = "student_habits_performance.csv"

# Load the latest version
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "jayaantanaath/student-habits-vs-academic-performance",
  file_path,
  # Provide any additional arguments like
  # sql_query or pandas_kwargs. See the
  # documenation for more information:
  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
)

# Drop rows with missing values in 'parental_education_level'
df.dropna(subset=['parental_education_level'], inplace=True)

average_study_hours = df['study_hours_per_day'].mean()
print(f"The average study hours per day for all students is: {average_study_hours:.2f} hours")
```

2. Identify the student who studies MOST hours per day. Please create a code cell below to answer the question.(0.5 point)

```python
student_most_study_hours = df.loc[df['study_hours_per_day'].idxmax()]
print("Student with the most study hours per day:")
print(student_most_study_hours[['student_id', 'study_hours_per_day']])
```

3. Count how many students study more than 6 hours per day. Please create a code cell below this to answer the question.(0.5 point)

```python
students_more_than_6_hours = df[df['study_hours_per_day'] > 6]
count_students_more_than_6_hours = len(students_more_than_6_hours)
print(f"Number of students who study more than 6 hours per day: {count_students_more_than_6_hours}")
```

4. What is the percentage of students who study more than 6 hours per day. Please create a code cell below this to answer the question.(0.5 point)

```python
total_students = len(df)
percentage_more_than_6_hours = (count_students_more_than_6_hours / total_students) * 100
print(f"Percentage of students who study more than 6 hours per day: {percentage_more_than_6_hours:.2f}%")
```

5. Calculate what percentage of students study less than 2 hours per day. Please create a code cell below this to answer the question.(0.5 point)

```python
students_less_than_2_hours = df[df['study_hours_per_day'] < 2]
count_students_less_than_2_hours = len(students_less_than_2_hours)
percentage_less_than_2_hours = (count_students_less_than_2_hours / total_students) * 100
print(f"Percentage of students who study less than 2 hours per day: {percentage_less_than_2_hours:.2f}%")
```

6. Do students who study more than 5 hours per day have higher exam scores on average? Please create a code cell below to answer this question. (0.5 point)

```python
high_study_students = df[df['study_hours_per_day'] > 5]
low_study_students = df[df['study_hours_per_day'] <= 5]

average_exam_score_high_study = high_study_students['exam_score'].mean()
average_exam_score_low_study = low_study_students['exam_score'].mean()

print(f"Average exam score for students studying more than 5 hours per day: {average_exam_score_high_study:.2f}")
print(f"Average exam score for students studying 5 hours or less per day: {average_exam_score_low_study:.2f}")

if average_exam_score_high_study > average_exam_score_low_study:
    print("\nStudents who study more than 5 hours per day have a higher average exam score.")
elif average_exam_score_high_study < average_exam_score_low_study:
    print("\nStudents who study more than 5 hours per day have a lower average exam score.")
else:
    print("\nThere is no significant difference in average exam scores between the two groups.")
```
